## Ekstraksi Data Teks (Bag-of-Words)
Mengubah data kalimat atau string menjadi representasi angka menggunakan teknik pembobotan frekuensi kata bernama `CountVectorizer`.

```python
from sklearn.feature_extraction.text import CountVectorizer

# 1. Menyiapkan Data Teks Toy Dataset
banned_words_reviews = [
    "The movie was great and fantastic",
    "I disliked the movie, it was bad and boring",
    "Great acting, but the story was bad"
]

# 2. Ekstraksi Fitur Teks dengan CountVectorizer
vect = CountVectorizer()
vect.fit(banned_words_reviews)

print("Vocabulary size:", len(vect.vocabulary_))
print("Vocabulary content:\n", vect.vocabulary_)

# Mengubah teks menjadi bentuk matriks angka (Sparse Matrix)
bag_of_words = vect.transform(banned_words_reviews)
print("Bag-of-words array representation:\n", bag_of_words.toarray())
```
