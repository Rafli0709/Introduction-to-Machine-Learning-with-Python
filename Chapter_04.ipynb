## Rekayasa Fitur
Melakukan konversi data kategori menggunakan One-Hot Encoding (`pandas.get_dummies`) serta otomatisasi seleksi fitur penting menggunakan `SelectPercentile`.

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectPercentile
from sklearn.linear_model import LogisticRegression

# 1. Simulasi Kategori Data (One-Hot Encoding)
demo_df = pd.DataFrame({'Target Value': [10, 20, 15, 30, 45],
                        'Categorical Feature': ['socks', 'fox', 'socks', 'box', 'fox']})
display(pd.get_dummies(demo_df))

# 2. Otomatisasi Fitur Seleksi (Feature Selection)
cancer = load_breast_cancer()
# Menambahkan fitur noise (tidak berguna) untuk simulasi seleksi
rng = np.random.RandomState(42)
noise = rng.normal(size=(len(cancer.data), 50))
X_w_noise = np.hstack([cancer.data, noise])

X_train, X_test, y_train, y_test = train_test_split(X_w_noise, cancer.target, random_state=0, test_size=.5)

# Memilih hanya 50% fitur terbaik
select = SelectPercentile(percentile=50)
select.fit(X_train, y_train)
X_train_selected = select.transform(X_train)

print("X_train shape with noise:", X_train.shape)
print("X_train shape after selection:", X_train_selected.shape)

# Perbandingan Model
lr = LogisticRegression(max_iter=10000)
lr.fit(X_train, y_train)
print(f"Score with all features: {lr.score(X_test, y_test):.3f}")
X_test_selected = select.transform(X_test)
lr.fit(X_train_selected, y_train)
print(f"Score with selected features: {lr.score(X_test_selected, y_test):.3f}")
```
