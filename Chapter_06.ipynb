## Chaining Workflow dengan Pipeline
Menggabungkan tahapan preprocessing (`StandardScaler`) dan algoritma pemodelan (`SVC`) menggunakan komponen `Pipeline` untuk mencegah terjadinya kebocoran data (*data leakage*).

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split, GridSearchCV

cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=0)

# 1. Membangun Pipeline Alur Kerja
pipe = Pipeline([("scaler", StandardScaler()), ("svm", SVC())])

# 2. Integrasi Pipeline ke dalam Grid Search
param_grid = {'svm__C': [0.01, 0.1, 1, 10, 100],
              'svm__gamma': [0.001, 0.01, 0.1, 1]}

grid = GridSearchCV(pipe, param_grid=param_grid, cv=5, n_jobs=-1)
grid.fit(X_train, y_train)

print(f"Best cross-validation accuracy: {grid.best_score_:.3f}")
print(f"Test set accuracy: {grid.score(X_test, y_test):.3f}")
```
