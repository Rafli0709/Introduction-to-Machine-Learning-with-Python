## Algoritma Utama & Contoh Dataset
Pada bab ini, kita mempraktikkan berbagai algoritma Supervised Learning (Klasifikasi dan Regresi) menggunakan dataset bawaan scikit-learn seperti Wisconsin Breast Cancer dan Boston Housing.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LinearRegression, Ridge, LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier

# 1. Memuat Dataset Kanker Payudara untuk Klasifikasi
cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(
    cancer.data, cancer.target, stratify=cancer.target, random_state=42
)

# 2. Eksperimen K-Nearest Neighbors (k-NN)
knn = KNeighborsClassifier(n_neighbors=6)
knn.fit(X_train, y_train)
print(f"k-NN - Train score: {knn.score(X_train, y_train):.3f}, Test score: {knn.score(X_test, y_test):.3f}")

# 3. Eksperimen Logistic Regression (Linear Model)
logreg = LogisticRegression(C=100, max_iter=10000)
logreg.fit(X_train, y_train)
print(f"Logistic Regression - Train score: {logreg.score(X_train, y_train):.3f}, Test score: {logreg.score(X_test, y_test):.3f}")

# 4. Eksperimen Decision Tree dengan Pre-pruning
tree = DecisionTreeClassifier(max_depth=4, random_state=0)
tree.fit(X_train, y_train)
print(f"Decision Tree - Train score: {tree.score(X_train, y_train):.3f}, Test score: {tree.score(X_test, y_test):.3f}")

# 5. Eksperimen Random Forest
forest = RandomForestClassifier(n_estimators=100, random_state=0)
forest.fit(X_train, y_train)
print(f"Random Forest - Train score: {forest.score(X_train, y_train):.3f}, Test score: {forest.score(X_test, y_test):.3f}")

# 6. Eksperimen Gradient Boosting
gbrt = GradientBoostingClassifier(random_state=0, max_depth=1)
gbrt.fit(X_train, y_train)
print(f"Gradient Boosting - Train score: {gbrt.score(X_train, y_train):.3f}, Test score: {gbrt.score(X_test, y_test):.3f}")

# 7. Eksperimen Support Vector Machine (SVM) dengan Scaling manual sederhana
min_on_training = X_train.min(axis=0)
range_on_training = (X_train - min_on_training).max(axis=0)
X_train_scaled = (X_train - min_on_training) / range_on_training
X_test_scaled = (X_test - min_on_training) / range_on_training

svm = SVC(C=1000, gamma='scale')
svm.fit(X_train_scaled, y_train)
print(f"SVM (Scaled) - Train score: {svm.score(X_train_scaled, y_train):.3f}, Test score: {svm.score(X_test_scaled, y_test):.3f}")

# 8. Eksperimen Multilayer Perceptron (Neural Network)
mlp = MLPClassifier(max_iter=1000, alpha=1, random_state=0)
mlp.fit(X_train_scaled, y_train)
print(f"Neural Network (Scaled) - Train score: {mlp.score(X_train_scaled, y_train):.3f}, Test score: {mlp.score(X_test_scaled, y_test):.3f}")
```
