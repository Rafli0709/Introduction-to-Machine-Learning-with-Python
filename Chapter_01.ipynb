## Penjelasan Teori Umum
Di bab ini, kita mempelajari dasar-dasar Machine Learning, perbedaan antara Supervised dan Unsupervised Learning, serta pentingnya data dalam membangun model.

```python
from sklearn.datasets import load_iris
iris_dataset = load_iris()
print("Keys of iris_dataset:", iris_dataset.keys())
```

### Penjelasan Dataset Iris:
Dataset Iris adalah dataset klasik dalam machine learning yang berisi 150 sampel dari 3 spesies bunga Iris (Setosa, Versicolor, dan Virginica). Setiap sampel memiliki 4 fitur: panjang sepal, lebar sepal, panjang petal, dan lebar petal.

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'], iris_dataset['target'], random_state=0)

knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
```

### Algoritma K-Nearest Neighbors (KNN):
KNN bekerja dengan cara mencari $k$ titik data terdekat dalam training set untuk menentukan label data baru berdasarkan mayoritas kelas terdekat tersebut. Di sini kita menggunakan $n\_neighbors=1$.

```python
print(f"Test set score: {knn.score(X_test, y_test):.2f}")
```
